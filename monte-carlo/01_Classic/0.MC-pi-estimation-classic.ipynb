# üßÆ Estimativa de $\pi$ usando o M√©todo de Monte Carlo

## Introdu√ß√£o

O m√©todo de Monte Carlo √© uma t√©cnica estat√≠stica baseada na gera√ß√£o de n√∫meros aleat√≥rios para resolver problemas matem√°ticos. Uma aplica√ß√£o cl√°ssica desse m√©todo √© a **estima√ß√£o do valor de $\pi$**.

A ideia central √© inscrever um c√≠rculo de raio $r = 1$ em um quadrado de lado $2r = 2$. Ao gerar pontos aleat√≥rios dentro do quadrado e contar quantos caem dentro do c√≠rculo, podemos estimar $\pi$ usando a propor√ß√£o entre os pontos dentro do c√≠rculo e o total de pontos gerados.

## Como funciona o c√°lculo?

Sabemos que:

- A √°rea do quadrado √©:  
  $$ A_{\text{quadrado}} = (2r) \times (2r) = 4 $$
- A √°rea do c√≠rculo de raio $r = 1$ √©:  
  $$ A_{\text{c√≠rculo}} = \pi r^2 = \pi $$

A raz√£o entre essas √°reas √©:  
$$ \frac{A_{\text{c√≠rculo}}}{A_{\text{quadrado}}} = \frac{\pi}{4} $$

Se gerarmos pontos aleat√≥rios $(x, y)$ dentro do quadrado e contarmos quantos deles caem dentro do c√≠rculo ($x^2 + y^2 \leq 1$), podemos estimar $\pi$ como:

$$ \pi \approx 4 \times \frac{\text{n√∫mero de pontos dentro do c√≠rculo}}{\text{total de pontos gerados}} $$

## Implementa√ß√£o do m√©todo

O c√≥digo abaixo:

1. Gera pontos aleat√≥rios $(x, y)$ no intervalo $[-1,1]$ para ambos os eixos.
2. Verifica se cada ponto est√° dentro do c√≠rculo.
3. Conta quantos pontos est√£o dentro do c√≠rculo e calcula $\pi$.
4. Exibe um gr√°fico com os pontos dentro (azul) e fora (vermelho) do c√≠rculo.
5. Exibe a estimativa de $\pi$.

Execute a c√©lula abaixo para ver o resultado!


```python
import random
import matplotlib.pyplot as plt

def estimate_pi_monte_carlo(num_points):
    inside_x, inside_y = [], []
    outside_x, outside_y = [], []
    
    inside_circle = 0  # Contador para os pontos dentro do c√≠rculo
    
    for _ in range(num_points):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        if x**2 + y**2 <= 1:
            inside_x.append(x)
            inside_y.append(y)
            inside_circle += 1
        else:
            outside_x.append(x)
            outside_y.append(y)
    
    # C√°lculo da estimativa de œÄ
    pi_estimate = 4 * inside_circle / num_points
    print(f"Estimativa de œÄ com {num_points} pontos: {pi_estimate}")
    
    # Plote os pontos
    plt.figure(figsize=(6, 6))
    plt.scatter(inside_x, inside_y, color="blue", s=1, label="Dentro do c√≠rculo")
    plt.scatter(outside_x, outside_y, color="red", s=1, label="Fora do c√≠rculo")
    plt.gca().set_aspect("equal")
    plt.legend()
    plt.title(f"M√©todo de Monte Carlo para \u03c0 com {num_points} pontos")
    plt.show()

# Testando com 10.000 pontos
estimate_pi_monte_carlo(10000)
```
